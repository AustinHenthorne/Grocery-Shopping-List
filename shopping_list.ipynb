```
# Author: Austin Henthorne
# Latest Update:   7.4.21
# Objective: Create a shopping list of ingredients by defining the 
#            dinners and meals the user would like.

# TO DO: Add GUI that is compatible with google colab

# Package all code into a function
def shopping_list():
    # Pick which meals you would like (True = yes, False = no)
    tikka_masala_flag        = False
    beef_burritos_flag       = False
    burgers_flag             = True
    chicken_fajitas_flag     = False
    eggplant_parm_flag       = False
    chicken_parm_flag        = False
    chicken_sandwhiches_flag = False
    beef_stew_flag           = False
    corned_beef_flag         = False
    turkey_nachos_flag       = False
    salmon_flag              = False
    sweet_potato_beef_flag   = False
    chicken_wings_flag       = False
    sausage_peppers_flag     = True
    beef_tacos_flag          = False
    chicken_teryaki_flag     = False
    steak_veg_flag           = False
    fish_veg_flag            = False
    Austin_WorkLunch_flag    = True
    Other_flag               = True

    file = open("/content/drive/MyDrive/GroceryList.txt", "w")
    grocery_list = []

    if tikka_masala_flag:
        tikka_masala = ["** Tikka Masala **", "Chicken Thighs", "Plain Yogurt", "Lemon",
                        "Pita Bread", "Coconut Milk"]
        grocery_list.extend(tikka_masala)
        print("Adding Tikka Masala ...")

    if beef_burritos_flag:
        beef_burritos = ["** Beef Burritos **", "1/2 lb Ground Beef", "Onion", "Rice",
                         "Avocado", "Sour Cream", "Tomato", "Wraps", "Tomato Sauce"]
        grocery_list.extend(beef_burritos)
        print("Adding Beef Burritos ...")

    if burgers_flag:
        burgers = ["** Burgers **", "1/2 lb Ground Beef", "Buns", "Tomato", "Avocado",
                   "Beans", "Spinach"]
        grocery_list.extend(burgers)
        print("Adding Burgers ...")

    if chicken_fajitas_flag:
        chicken_fajitas = ["** Chicken Fajitas **", "1 Chicken Breast", "2 Peppers", "1 Onion"]
        grocery_list.extend(chicken_fajitas)
        print("Adding Chicken Fajitas ...")

    if eggplant_parm_flag:
        eggplant_parm = ["** Eggplant Parmesian **", "1 Eggplant", "Sauce", "Mozzarella", "Breadcrumb", "Egg"]
        grocery_list.extend(eggplant_parm)
        print("Adding Eggplant Parmesian ...")

    if chicken_parm_flag:
        chicken_parm = ["** Chicken Parmesian **", "2 Chicken breasts", "Sauce", "Mozzarella", "Breadcrumb", "Egg"]
        grocery_list.extend(chicken_parm)
        print("Adding Chicken Parmesian ...")

    if chicken_sandwhiches_flag:
        chicken_sandwhiches = ["** Chicken Sandwiches **", "Chicken Breasts", "Buns", "Tomato"]
        grocery_list.extend(chicken_sandwhiches)
        print("Adding Chicken Sandwiches ...")

    if beef_stew_flag:
        beef_stew = ["** Beef Stew **", "1 lb Stew Meat (chuck roast)", "3 Russet Potatos", "4 Large Carrots",
                     "20 oz Beef Broth", "1 Onion", "1 Can Tomato Sauce"]
        grocery_list.extend(beef_stew)
        print("Adding Beef Stew ...")

    if corned_beef_flag:
        corned_beef = ["~3lb Corned Beef", "Small Red Potatoes", "5 Carrots", "Cabbage","Rye Bread",
                       "Thousand Island"]
        grocery_list.extend(corned_beef)
        print("Adding Corned Beef ...")

    if turkey_nachos_flag:
        turkey_nachos = ["** Turkey Nachos **", "Ground Turkey", "Black beans", "Black olives", "Nacho cheese",
                         "Chips", "Avocado", "Red Onion", "Corn", "Tomato Sauce"]
        grocery_list.extend(turkey_nachos)
        print("Adding Turkey Nachos ...")

    if salmon_flag:
        salmon = ["** Salmon **", "Salmon Fillet", "Cauliflower"]
        grocery_list.extend(salmon)
        print("Adding Grilled Salmon ...")

    if sweet_potato_beef_flag:
        sweet_potato_beef = ["2 Medium Sweet Potatos", "1/2 lb Ground Beef", "Avocado",
                             "Sour Cream"]
        grocery_list.extend(sweet_potato_beef)
        print("Adding Sweet Potato with Ground Beef...")

    if chicken_wings_flag:
        chicken_wings = ["** Chicken Wings **", "2 lbs chicken wings", "Beans", "Bbq sauce"]
        grocery_list.extend(chicken_wings)
        print("Adding Chicken wings with beans...")

    if sausage_peppers_flag:
        sausage_peppers = ["** Sausages and Peppers **", "12 oz Italian sausage", "Yellow & Red pepper", "1 onion",
                           "Red wine vinegar", "Bread"]
        grocery_list.extend(sausage_peppers)
        print("Adding Sausage and Peppers...")

    if beef_tacos_flag:
        beef_tacos = ["** Beef Tacos **", "1/2 lb Ground beef", "Shells", "1 Tomato", "1 Avocado",
                      "Tomato sauce"]
        grocery_list.extend(beef_tacos)
        print("Adding Beef tacos...")

    if chicken_teryaki_flag:
        chicken_teryaki = ["** Chicken Teryaki **", "1 lb Chicken Breasts", "Honey", "Brown Sugar", "Sesame Oil"]
        grocery_list.extend(chicken_teryaki)
        print("Adding Chicken Teryaki...")

    if steak_veg_flag:
        steak_veg = ["** Steak & Vegs **", "Steak", "Vegetable", "Beans"]
        grocery_list.extend(steak_veg)
        print("Adding Steak and Vegs...")

    if fish_veg_flag:
        fish_veg = ["** Fish & Vegs **", "Fish (Salmon or Mahi Mahi)", "Vegetable"]
        grocery_list.extend(fish_veg)
        print("Adding Fish and Vegs...")

    if Austin_WorkLunch_flag:
        Austin_WorkLunch = ["** Austin Work Lunch **", "Carrots", "Blueberries", "Strawberries", "Sandwich Bread",
                            "Chomps", "Pretzels", "Oatmeal Cups"]
        grocery_list.extend(Austin_WorkLunch)
        print("Adding Austins Work Lunch ...")

    if Other_flag:
        other = ["** Misc. **", "Apples", "Oranges", "Greens (Spinach)", "Pepper", "Cucumber",
                 "Turkey", "Nuts"]
        grocery_list.extend(other)
        print("Adding Other ...")

    with file as f:
        for item in grocery_list:
            f.write("%s\n" % item)

# Main ()
shopping_list()
print("Complete.")
```
